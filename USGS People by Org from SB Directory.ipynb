This notebook shows a really simple example of pulling information from the ScienceBase Directory by FBMS OrgCode (BCB org code as an example). This might be a reasonable way to put together a full directory for USGS staff and locations. We can start from the perspective of a fresh data dump of FBMS codes and the few associated properties and then walk the directory to pull the few fields on active USGS staff we want, including city/state or zip codes to geolocate and map. One interesting thing that this example shows is the inclusion of Anne Davidson, a USGS cooperator with Boise State University who does hold Active Directory credentials. That will likely be the case for other orgs, throwing in some data that goes beyond direct USGS employees.


```python
import requests
from IPython.display import display
```

```python
sbURL = "https://www.sciencebase.gov/directory/people?q=GGHIEJ4000&format=json&dataset=all&max=10&offset=0&active=true"
dirData = requests.get(sbURL).json()

for person in dirData["people"]:
    print (person["displayName"])
    if "jobTitle" in person:
        print (person["jobTitle"])
    print (person["primaryLocation"]["streetAddress"]["city"]+", "+person["primaryLocation"]["streetAddress"]["state"], person["primaryLocation"]["streetAddress"]["zip"])
    print ("----------")
```
